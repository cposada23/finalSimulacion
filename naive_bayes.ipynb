```python
from __future__ import division
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
#Algunas advertencias que queremos evitar
import warnings
warnings.filterwarnings("always")
%matplotlib inline
```

```python
#cargamos la bd de entrenamiento
db = pd.read_csv('Postures.csv',delimiter=',')  # Assuming tab-delimiter
db.head()
print ("Dim de la base de datos original:" + str(np.shape(db)))
DataBase = db
```

```python
DataBase = DataBase.replace({'?': 1})
```

```python
DataBase
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import urllib
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.cross_validation import train_test_split
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from matplotlib.colors import ListedColormap
%matplotlib inline
```

```python
X=DataBase.iloc[0:-1, 2:-1].values
y=DataBase.iloc[0:-1, 0].values
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=0)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
clasificador = GaussianNB()
clasificador.fit(X_train, y_train)
y_pred=clasificador.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
import seaborn as sn
df_cm = pd.DataFrame(cm, index = [i for i in "12345"],
                  columns = [i for i in "12345"])
plt.figure(figsize = (15,7))
sn.heatmap(df_cm, annot=True)
```
