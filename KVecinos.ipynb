```python
# Importación de librerias 

from __future__ import division
import numpy as np
from numpy import random
import scipy as sc

from sklearn.model_selection import train_test_split
import pandas as pd
import math
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import mode
import operator
#Algunas advertencias que queremos evitar
import warnings
warnings.filterwarnings("always")
%matplotlib inline
```

```python
#cargamos la bd de entrenamiento
db = pd.read_csv('Postures.csv',delimiter=',')  # Assuming tab-delimiter
db.head()
print ("Dim de la base de datos original:" + str(np.shape(db)))
DataBase = db
DataBase = DataBase.replace({'?': 1})
DataBase
```

```python
X = DataBase.iloc[2:, 2:]
Y = DataBase.iloc[2:, 0]
print("Tamaño de X: " , X.shape , " Tamaño de Y: ", Y.shape)
```

```python
#Para calcular el error en los problemas de clasificación
def ErrorClas(Y_lest, Y):
    error = 1 - np.sum(Y_lest == Y)/len(Y)
    
    return error
```

```python
## Metodo de KNN para Clasificación

def KNN(X_train, Y_train, X_val, k, tipo):
    
    
    tipo = tipo
    k = k    #Parámetro k que equivale al número de vecinos a tener en cuenta para resolver el problema de
               #predicción de la variable de salida
    N = X_train.shape[0]
    M = X_val.shape[0]
    
    dist = np.zeros((N,M))
    L = np.dot(X_train, X_val.T)
    
    tem = np.sum(X_train*X_train, axis=1).reshape([X_train.shape[0], 1])
    tem2 = np.sum(X_val*X_val, axis=1).reshape([1, X_val.shape[0]])
    dist = np.dot(tem, np.ones((1,X_val.shape[0]))) + np.dot(np.ones((X_train.shape[0], 1)), tem2) - 2*L
    dist = np.sqrt(dist)
    dist = dist.T
    otra = np.argsort(dist, axis = 1)
    otra = otra[:, 0:k]
    if(tipo == "regresion"):
        return np.mean( Y_train[otra], axis = 1)
    elif(tipo == "clasificacion"):
        return mode(Y_train[otra], axis=1)[0] 
```

```python
## Programa principal

N = np.size(X,0)

#Validamos el modelo
random.seed(1)
Error = np.zeros(4)
k = 7
tipo = "clasificacion"
for j in range(4):
    print("Working")
    
    Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.25)

    #Normalizamos los datos
    media = np.mean(Xtrain)
    desvia = np.std(Xtrain)
    Xtrain = sc.stats.stats.zscore(Xtrain)
    Xtest = (Xtest - np.matlib.repmat(media, Xtest.shape[0], 1))/np.matlib.repmat(desvia, Xtest.shape[0], 1)

    Yest = KNN(Xtrain, Ytrain, Xtest, k, tipo)#Complete con el llamado apropiado de la función que Uds diseñaron
#     print(Yest)
    #Evaluamos las predicciones del modelo con los datos de test
    Error[j] = ErrorClas(Yest.T,Ytest)
print('\nError durante la prueba = ' + str(np.mean(Error)) + '+-' + str(np.std(Error)))
```
